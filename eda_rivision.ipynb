P**earson Correlation of non categorical data**

```python
df_cleaned.astype(int)
```

```python
from scipy.stats import pearsonr

# list of features to check against target
pearson_corr={}
select_feature=['age','bmi','children','is_smoker','region_northwest','region_southeast','region_southwest','bmi_cat_Normal weight','bmi_cat_Overweight','bmi_cat_obese'];
for i in select_feature:
  pearson_corr[i]=pearsonr(df_cleaned[i],df_cleaned['charges'])[0]

correlation_df = pd.DataFrame(list(pearson_corr.items()), columns=['Feature', 'Pearson Correlation'])

correlation_df

```

```python
cat_features = [
    'isfemale', 'is_smoker',
    'region_northwest', 'region_southeast', 'region_southwest',
    'bmi_cat_Normal weight', 'bmi_cat_Overweight', 'bmi_cat_obese'
]
```

```python
from scipy.stats import chi2_contingency
import pandas as pd
alpha=0.05
# bin the target variable
df_cleaned['charges_bin']=pd.qcut(df['charges'],q=4 ,labels=False)

# Run Chi-square test
chi2_result={}

for i in cat_features:
  contingency=pd.crosstab(df_cleaned[i],df_cleaned['charges_bin'])
  chi2_stat,p_val,_,_=chi2_contingency(contingency)
  decision='Reject NULL (Keature)' if p_val < alpha else 'Accept (Drop feature)'
  chi2_result[i]={'chi2_stats':chi2_stat,
                  'p_val':p_val,
                  "Desicion":decision}
chi2_df=pd.DataFrame(chi2_result).T
chi2_df
```
