```python
col=['discounted_price','quantity_sold','price','total_revenue']
```

```python
finaldf=df[col]
finaldf.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
finaldf=pd.DataFrame(scale.fit_transform(finaldf),columns=finaldf.columns)
finaldf
```

```python
X=finaldf.drop('total_revenue',axis=1)
y=finaldf['total_revenue']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(X_train,y_train)
```

```python
y_pred=le.predict(X_test)
```

```python
from sklearn.metrics import r2_score
val=r2_score(y_test,y_pred)
val
```
